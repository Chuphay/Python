```python
#"batch" Gradient Descent
def linGet(x,y,alpha,theta = [0,1]):
    L = len(x)
    out = theta[:]
    for i in range(L):
        out[0] += alpha*(theta[0] +theta[1]*x[i] - y[i])/L
        out[1] += alpha*(theta[0] +theta[1]*x[i] - y[i])*x[i]/L
    return out   
line = lambda x, ls = [0,1]: ls[0]+ls[1]*x
x = linspace(0,10,101)
y = x+1.5*randn(101)
xlim(0,10)
ylim(-2,12)
plot(x,y,'m.')
tt = linGet(x,y,-0.02,[3,-1])
plot(x,line(x,[3,-1]),'r')
for i in range(1000):
    tt = linGet(x,y,-0.05,tt)
    if i%50==0:
        plot(x,line(x,tt))
plot(x,line(x,tt),'k')    
print tt    
```
