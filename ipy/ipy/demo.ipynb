```python
from scipy.fftpack import fft

L = 10
n = 2048
t2 = np.linspace(0,L,n+1)
t = t2[:n]
k = 2*pi/L*np.append(np.arange(n/2),np.arange(-n/2, 0))
ks = fftshift(k)
S = 6*sin(38*t)*exp(-t)-2*exp(-(t-3)**2)+2*cos(2*t)-(t-2)*(t-5)/6.5+4*sin(10*t)*exp(-t/3)+3*cos(15*t)*exp(-(t-7)**2) -4*sin(2.5*t)
figsize(20,15)

slide = linspace(0,10,101)
width = 10

data = []
for s in slide:
    f = exp(-width*(t-s)**2)
    sf = S*f
    sft = fft(sf)

    subplot(3,1,2), plot(t,sf), plot(t,S,'k-')
    subplot(3,1,3), plot(ks,(fftshift(sft))), xlim(-30,30)
    data.append(abs(fftshift(sft)))
```
