```python
def repeat(times):
    """Modified from: http://www.artima.com/weblogs/viewpost.jsp?thread=240845
    For example:   
    
    @repeat(1000000)
    def getRand():
        return np.random.rand(3)
    print getRand()
    
    #should return something like array([0.5,0.5,0.5])
    """
    def wrap(f):
        def wrapped_f(*args):
            output = f(*args)
            for i in range(times-1):
                output += f(*args)
            return output/float(times)
        return wrapped_f
    return wrap
```

```python
@repeat(10000)
def roll2():
    return randint(1,7)+randint(1,7)
print roll2()

@repeat(10000)
def flip_till_heads():
    i = 1
    while(rand()<0.5):
        i += 1
    return i
print flip_till_heads()
```

```python
prob = 0.5
size = 15
figsize(12,8)
seed(123)
"""
for i in range(size):
    for j in range(size):
        scatter(i,j,c='b')
"""
plot_data = []
data = {key: [] for key in range(1,(size)**2+1)}
for i in range(size):
    for j in range(size):
        scatter(i,j,c='b')
        if(j<(size-1) and rand()>prob):
            plot_data.append([j,j+1])
            plot_data.append([i,i])
            plot_data.append('k')
            data[i*size+j+1].append(i*size+j+2)
            data[i*size+j+2].append(i*size+j+1)
        if(i<(size-1) and rand()>prob):    
            plot_data.append([j,j])
            plot_data.append([i,i+1])
            plot_data.append('k')
            data[i*size+j+1].append((i+1)*size+j+1)
            data[(i+1)*size+j+1].append(i*size+j+1)
plot_data = array(plot_data)
plot(*plot_data);    

visited = {key: False for key in range(1,(size)**2+1)}
def DFS(data, node): 
    scatter((node-1)%size,(node-1)/size, c='r',s=30,lw=0)
    visited[node] = True
    for i in range(len(data[node])):
        new_node = data[node][i]
        plot([(node-1)%size,(data[node][i]-1)%size],[(node-1)/size,(data[node][i]-1)/size],'r',linewidth=1.5)
        if(visited[new_node]==False):
            DFS(data, new_node)

DFS(data,10) 
```
