#### Uniform Random Numbers

'There is no such thing as <em>random</em> in a computer program. A computer will always produce the same result if the input is the same. A random-number generator here really means a pseudo-random-number generator that can generate a long sequence of numbers that can imitate a given distribution.



http://en.wikipedia.org/wiki/Random_Number_Generator

Pseudo-random number generators (PRNGs) are algorithms that can automatically create long runs of numbers with good random properties but eventually the sequence repeats (or the memory usage grows without bound). The string of values generated by such algorithms is generally determined by a fixed number called a seed. One of the most common PRNG is the linear congruential generator, which uses the recurrence

$X_{n+1} = (a X_n + b)\, \textrm{mod}\, m$

to generate numbers. The maximum number of numbers the formula can produce is the modulus, m.

Where $a$, $b$, $m$ are <em>magic numbers</em>: their values determine the quality of the generator. One common choice, $a=7^5=16807, b = 0,$ and $c=2^{31}-1=2147483647$ 

```python
from datetime import datetime
now = datetime.now()
seed = now.year + 70*(now.month + 12*(now.day + 31*(now.hour + 23*(now.minute +59*now.second))))

def ranf(*args):

    if len(args) == 0:
        global seed
        x = seed
    else:
        x = args[0]
    a, c, q, r = 16807, 2147483647, 127773, 2836 
    h, l = x/q, x%q
    t = a*l - r*h
    if t>0:
        seed = t
        return float(t)/c
    else:
        seed = c+t            
        return float(c+t)/c 
```

Note that in the above code, we have used two more magic numbers $q=c/a$ and $r = c \, \textrm{mod}\,a$. Park and Miller implemented this in 1988, to show that the method given here does implement the algorithm correctly, we can se the initial seed to be 1, and then after 10,000 steps, we should have 1,043,618,065 returned as the value of the seed. 

```python
seed = 1
for i in xrange(10000):
    x = ranf(seed)   
seed == 1043618065
```

As soon as we obtain good unifor random-number generators, we can use them to create other types of random-number generators. For example, we can use a uniform random-number generator to create an exponential distribution or a Gaussian distribution.

One way to generate the exponential distribution is to relate it to a uniform distribution. For example, if we have a uniform distribution $f(y)=1$ for $ y \in [0,1]$, we can relate it to an exponential distribution by

$$f(y)dy=dy=p(x)dx = e^{-x}dx$$

which gives

$$y(x) - y(0) = 1 - e^{-x}$$

after integration. We can set $y(0) = 0$ and invert the above equation to have

$$x = -\ln(1-y),$$

which relates the exponential distribution of $x \in [0,\infty)$ to the uniform distribution of $y \in [0,1]$.

```python
def rane(*args):
           
    if len(args) == 0:
        global seed
        x = seed
    else:
        x = args[0]
    return -log(1-ranf(x))   
```

Another useful distribution in physics is the Gaussian distribution

$$g(x) = \frac 1 {\sqrt {2 \pi \sigma}} e^{-x^2/2\sigma^2}$$

where $\sigma$ is the variance of the distribution, which we can take as $1$ for the moment. The distribution woth $\sigma \neq 1$ can be obtained via the rescaling of $x$ by $\sigma$. 

We can use a uniform distribution $f(\phi) = 1$ for $\phi \in [0,2\pi]$ and an exponential distribution $p(t) = e^{-t}$ for $t \in [0, \infty)$ to obtain two Gaussian distributions $g(x)$ and $g(y)$. We can relate the product of a uniform distribution and an exponential distribution to a product of two Gaussian distributions by

$$\frac 1 {2 \pi} f(\phi)d \phi ~p(t) dt = g(x) dx ~g(y) dy $$

which gives

$$e^{-t}dtd \phi = e^{-(x^2 + y^2)/2}dxdy$$

The above equation can be viewed as the coordinate transform from the polar system $(\rho , \phi )$ with $\rho = \sqrt {2t} $ into the rectangular system $(x,y)$, that is,

$$x = \sqrt {2t} \cos \phi $$
$$y = \sqrt {2t} \sin \phi $$

which are two Gaussian distributions if $t$ is taken from an exponential distribution and $\phi$ is taken from a uniform distribution in the region $[0,2 \pi ]$. 

```python
def rang(*args):
              
    if len(args) == 0:
        global seed
        x = seed
    else:
        x = args[0]
    rdm1 = - log(1-ranf(x))
    rdm2 = 2*pi*ranf(seed)
    return rdm1*cos(rdm2)  
```

```python
figsize(14,2)

x = [ranf() for i in range(10000)]
x,y = histogram(x, bins =100, range=(-2,3))
subplot(131), plot(y[:len(x)],x/10000.)

y = [rane() for i in range(10000)]
x,y = histogram(y, bins =100, range=(-1,6))
subplot(132), plot(y[:len(x)],x/10000.)

z = [rang() for i in range(10000)]
x,y = histogram(z, bins =100, range= (-3,3))
subplot(133),plot(y[:len(x)],x/10000.) #,axis('off') 

show()
```
