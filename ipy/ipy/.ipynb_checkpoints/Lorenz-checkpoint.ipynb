```python
from scipy.integrate import odeint

def lorenz(y,t):
    x, y, z = y[0],y[1],y[2]
    sigma, ro, beta = 10, 28, (8.0)/3
    dx_dt = sigma*(y - x)
    dy_dt = x*(ro - z) - y
    dz_dt = x*y - beta*z
    return [dx_dt, dy_dt, dz_dt]

t = linspace(0,6,1000)

z = odeint(lorenz,[10,0.1,6,7],t)

plot(z[:,2],z[:,0]);
```
