```python
from scipy.integrate import odeint

def lorenz(y,t):
    x, y, z = y[0],y[1],y[2]
    sigma, ro, beta = 10, 28, (8.0)/3
    dx_dt = sigma*(y - x)
    dy_dt = x*(ro - z) - y
    dz_dt = x*y - beta*z
    return [dx_dt, dy_dt, dz_dt]

t = linspace(0,6,1000)

z = odeint(lorenz,[10,0.1,6,7],t)

plot(z[:,2],z[:,0]);
```

```python
def get_next(x, step):
    new_position = x[0] + step*x[1]
    new_velocity = x[1] - step*x[0]
    return [new_position, new_velocity]

z = [[0,1]]
for i in range(1000): 
    z.append(get_next(z[i],0.01))
z = array(z)    
plot(z[:,0]); 
```

```python
data = []
r = linspace(0.8,1.3,100)
for step in r:
    z = [0,1]
    for i in range(50): 
        z = get_next(z,step)   
    data.append(z)


data = array(data) 
print len(data) , len (r)
scatter(r,data[:,0])
show()
```
