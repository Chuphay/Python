```python
import scipy as sp
from scipy import exp,pi
from scipy.integrate import quad
def fourier(function,k):
    """hmmm... assuming the range is from 0 to 2*pi.
    function should be some kind of lambda like function
    k is the kth coefficient, which is supposedly what will be returned to you 
    """
    fn = lambda x: function(x)*exp(1j*k*x)
    real_part = lambda x: sp.real(fn(x))
    imag_part = lambda x: sp.imag(fn(x))
    return 1/(2*pi)*quad(real_part,0,2*pi)[0]+1j/(2*pi)*quad(imag_part,0,2*pi)[0]
def transform(function,a,b):
    """this should translate any function, so that it is useable by
    the fourier transform above"""
    out = lambda x: function((b-a)*x/(2*pi)+a)
    return out
y = lambda x : x/(2*pi)
for i in range(6):
    print fourier(y,i)
print sp.around(fourier(y,2),4)
z = lambda x: x
z_prime = transform(z,0,1)
print z_prime(2*pi)
print sp.around(fourier(z_prime,2),4)  
```

```python
from scipy.integrate import quad
def fourier(function,k,a=0,b=2*pi):
    """hmmm... assuming the range is from 0 to 2*pi.
    function should be some kind of lambda like function
    k is the kth coefficient, which is supposedly what will be returned to you 
    """
    fn = lambda x: function((b-a)*x/(2*pi)+a)*exp(1j*k*x)
    real_part = lambda x: real(fn(x)) #sp.real()
    imag_part = lambda x: imag(fn(x))
    return 1/(b-a)*quad(real_part,a,b)[0]+1j/(b-a)*quad(imag_part,a,b)[0]


def step(an_array, where=3):
    out = ones(len(an_array))
    for i in range(len(an_array)):
        if an_array[i]>=where:
            break
        else:
            out[i] = -1*out[i]
  
    return out


n = 2048

x = np.linspace(0,2*pi,n)


plot (x,step(x,pi))
ylim(-2,2)
```
