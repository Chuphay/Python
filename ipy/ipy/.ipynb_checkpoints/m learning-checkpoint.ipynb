```python
J = lambda theta:4+(theta-2)**2
x = linspace(0,5,101)
plot(x,J(x),'k')
def get_new(theta,alpha):
    return theta - alpha*(J(theta+0.1)-J(theta))/0.1
theta_0 = 4.5
plot(theta_0,J(theta_0),'ro')
theta_1 = get_new(theta_0,0.8)
plot([theta_0,theta_1],[J(theta_0),J(theta_1)],'r',theta_1,J(theta_1),'ro')
theta_2 = get_new(theta_1,0.8)
plot([theta_1,theta_2],[J(theta_1),J(theta_2)],'r',theta_2,J(theta_2),'ro')
theta_3 = get_new(theta_2,0.8)
plot([theta_2,theta_3],[J(theta_2),J(theta_3)],'r',theta_3,J(theta_3),'ro')
theta_4 = get_new(theta_3,0.8)
plot([theta_3,theta_4],[J(theta_3),J(theta_4)],'r',theta_4,J(theta_4),'ro')
ylim(3,14)
show()
    
    
```
