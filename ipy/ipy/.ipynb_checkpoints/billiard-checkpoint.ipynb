```python
#we assume that we started ipython in pylab mode
%pylab inline
#here we set up our global variables
speed = 10.0
amount_of_crazy = 0.2
seed(123)
wall = 10.0
x = array([0.0,wall])

#here we get ready to make some data
data = []
for i in range(1000):
    theta = amount_of_crazy*randn()
    velocity = speed*array([cos(theta),sin(theta)])*(1+amount_of_crazy*randn())

    time = wall/velocity[0]
    y = array([0.0,velocity[1]*time])
    if i<20:
        plot(x,y)
    #the only two pieces of interesting data are the time and the final y-position
    #so we will add that to our data set
    data.append([time,y[1]])
xlabel('x-axis')
ylabel('y-axis')
print data[:5]
```

```python
from pandas import DataFrame
df = DataFrame(data)
df.columns=['time','y-displacement']

#df.describe()
#(((df['y-displacement']**2+100)**0.5)/df['time']).plot()
df['time'].hist(bins =100,histtype = 'step')
#show()
df.describe()
#df.hist?
#http://stackoverflow.com/questions/20198538/poisson-distribution-fit
from scipy.misc import factorial
from scipy.optimize import curve_fit
z,x=np.histogram(df['time'],bins=100)
x = x[:z.size]
fitfunc = lambda x,a,b,c: a*(b**(c*x))*(exp(-b))/factorial(c*x)
pars, covmat = curve_fit(fitfunc,x,z)

plt.plot(x, fitfunc(x, *pars), 'r-')


```

```python
from scipy.stats import poisson
mu = arange(0,100)
rv = poisson(mu)
x = np.arange(0, np.minimum(rv.dist.b, 100))
h = plt.vlines(x, 0, rv.pmf(x), lw=2)
```

```python
#df['y-displacement'].hist(bins=100)
#http://nbviewer.ipython.org/gist/keflavich/4042018
from scipy.optimize import curve_fit

z,x=np.histogram(df['y-displacement'],bins=100)
x = x[:z.size]
plot(x,z)
gaussfunc = lambda x,a,b,c,d: a*np.exp(-((x-b)/c)**2)+d
fitpars, covmat = curve_fit(gaussfunc,x,z)


plt.plot(x, gaussfunc(x, *fitpars), 'r-')

#http://stackoverflow.com/questions/7805552/fitting-a-histogram-with-python
from scipy.optimize import leastsq
fitfunc  = lambda p, x: p[0]*exp(-0.5*((x-p[1])/p[2])**2)+p[3]
errfunc  = lambda p, x, y: (y - fitfunc(p, x))
init  = [1.0, 0.5, 0.5, 0.5]
out   = leastsq( errfunc, init, args=(x, z))
c = out[0]
plot(x, fitfunc(c, x),'k-')
```

```python
df['distance'] = (df['y-displacement']**2+100)**0.5
df['speed'] = df['distance']/df['time']
df['speed'][:40].plot(marker='o')

from scipy.optimize import curve_fit
linear = lambda x,a,b: a*x+b
x = arange(40)
fitpars, covmat = curve_fit(linear,x,df['speed'][x])
plt.plot(x, linear(x, *fitpars), 'r-')

```

```python
from pandas import DataFrame,Series
d1 = DataFrame( {'StudentID': ["x1", "x10", "x2","x3", "x4", "x5", "x6",   "x7",     "x8", "x9"],
                       'StudentGender' : ['F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'M', 'M'],
                 'ExamenYear': ['2007','2007','2007','2008','2008','2008','2008','2009','2009','2009'],
                 'Exam': ['algebra', 'stats', 'bio', 'algebra', 'algebra', 'stats', 'stats', 'algebra', 'bio', 'bio'],
                 'Participated': ['no','yes','yes','yes','no','yes','yes','yes','yes','yes'],
                  'Passed': ['no','yes','yes','yes','no','yes','yes','yes','no','yes']})
d1['testValue'] = 80+10*randn(len(d1))
def ZahlOccurence_1(x):
    return Series({'All': len(x['StudentID']),
        'Part': sum(x['Participated'] == 'yes'),
        'Pass' :  sum(x['Passed'] == 'yes'),
        'test_mean' : x['testValue'].mean(),
        'test_std' : x['testValue'].std()})


d1.groupby('ExamenYear').apply(ZahlOccurence_1)
```
