#### Differential Equations

Ordinary differential equations are pretty important. In this blog post I'm going to try to learn how to do them numerically. 

I'm going to be referencing a book called Computational Physics by Tao Pang for this post.

The first, probably most easiest, thing to examine is a mass under an elastic force (i.e. a mass attached to a spring).

Let's write out Newton's equation (Hooke's law) for this:

$$F_{orce} = mass*acceleration = -kx$$

where k is the spring constant. To make our lives easy, let's just say that $k=mass=1$. Then we can rewrite this equation as

$$a = \frac {dv}{dt} = -x $$

where $v = dx/dt$.

```python
def get_x(old_x,velocity,time):
    return float(old_x + velocity*time)
def get_velocity(old_velocity,acceleration,time):
    return float(old_velocity + acceleration*time)
x = [0]
v = [1]
t = [0]
step = pi/50
for i in range(100): 
    t.append(t[i]+step)
    x.append(get_x(x[i],v[i],step))
    v.append(get_velocity(v[i],-x[i],step))
             
plot(t,x,label ='numerical'),plot(t,sin(t),'k',label='analytical')
legend()
show()
```

This is okay, but let's see what happens over a longer period of time.

```python
x,v,t = [0],[1],[0]
for i in range(1000): 
    t.append(t[i]+step)
    x.append(get_x(x[i],v[i],step))
    v.append(get_velocity(v[i],-x[i],step))
             
plot(t,x,label ='numerical'),plot(t,sin(t),'k',label='analytical')
legend(loc='best')
show()
```

Here we can see that the numerical approximation is getting worse and worse. Let's see if we can do better.

```python
from scipy.integrate import odeint

f = lambda y,t: [y[1],-y[0]]
t  = linspace(0, 24*pi, 200)   
x = odeint(f, [0,1], t)[:,0]

plot(t,x,label ='numerical'),plot(t,sin(t),'k',label='analytical')
legend(loc='best')
show()

```

```python
from scipy.integrate import odeint

def f(y,t): 
    x_vel = y[0]
    y_vel = y[1]
    x_pos = y[2]
    y_pos = y[3]
    x_vel = x_pos/((x_pos**2+y_pos**2)**1.5)
    y_vel = y_pos/((x_pos**2+y_pos**2)**1.5)
    x_pos = y[0]
    y_pos = y[1]
    return [x_vel,y_vel,x_pos,y_pos]



t  = linspace(0, 10, 400) 
data = []
for i in range(41):
    
    soln = odeint(f, [1,0,-5,10-i/2.0], t)
    if soln[-1][2] >= 3:
        data.append([10-i/2.0,soln[-1][2],soln[-1][3]])
    plot(soln[:,2],soln[:,3])
plot(0,0,'k.')
show()
```

So, we are trying to show that the differential cross section is $\sigma (\theta) = \frac 1 {16E^2\sin^4(\frac \theta 2)}$ where E is the energy

the differential cross section $\sigma (\theta)$ is the probability of a particle being found in the solid angle element $d\Omega = 2\pi\sin\theta d\theta$ at the deflection angle $\theta$

"The differential scattering cross section is defined as the number of interactions per target particle that lead to scattering into an element of the solid angle at a given angle divided by the number of incident particles per unit area."

$\sigma (\theta) = -\frac b {sin(\theta)} \frac {db}{d\theta}$

and then we are looking to plot $\ln \sigma (\theta) $ vs. $\theta$

Let's just try that first

```python
def sigma(theta, energy=1):
    denom = (16*energy**2)*(sin(theta/2))**4
    return 1/denom

x = linspace(0.1,1,100)
plot(x,log(sigma(x)))
```
