```python
#http://nbviewer.ipython.org/github/c3h3/Computational-Methods-for-Data-Analysis/tree/master/
import numpy as np
import matplotlib.pyplot as plt

L = 10
n = 2048
t2 = np.linspace(0,L,n+1)
t = t2[:-1]
k = (2*np.pi/L)*np.fft.fftfreq(n,1.0/n)
ks = np.fft.ifftshift(k)

S = (3*np.sin(2*t) + 0.5*np.tanh(0,5*(t-3)) + 0.28*np.exp(-(t-4)**2) \
     + 1.5*np.sin(5*t) + 4*np.cos(3*(t-6)**2))/10 + (t/20)**3

width = 1
slides = np.arange(0,10,0.1)
spec = np.array([])
spec_n = np.array([])
for i in range(len(slides)):
    f = np.exp(-width*(t-slides[i])**2)
    Sf = np.prod(np.c_[f,S],axis=1)
    Sft = np.fft.fft(Sf)

    if i == 0:
        spec = np.r_[spec,np.abs(np.fft.fftshift(Sft))]
        #spec_n = np.r_[spec_n,np.abs(np.fft.fftshift(Sft))/max(np.abs(np.fft.fftshift(Sft)))]
    else:
        spec = np.c_[spec,np.abs(np.fft.fftshift(Sft))]

print len(slides), len(ks), len(spec)
plt.pcolor(slides,ks,spec,  cmap="hot")
plt.ylim(-50, 50)
plt.colorbar()
plt.xlabel("t")
plt.ylabel("omega")
plt.show()
```
