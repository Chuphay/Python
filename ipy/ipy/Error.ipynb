```python
%pylab inline
#http://nbviewer.ipython.org/gist/keflavich/4042018
x = linspace(0,10,50)
y = 2*x + 1.2
plot(x,y)
noise = randn(y.size)
z = y + 4*noise
plot(x,z,marker='o')
# We'll use polyfit to find the values of the coefficients.  The third
# parameter is the "order"
p = np.polyfit(x,z,1)
plot(x,p[0]*x+p[1],'r-')
p = polyfit(x,z,2)
plot(x,p[0]*x**2+p[1]*x+p[2],'k')
```

```python
#http://nbviewer.ipython.org/gist/keflavich/4042018
x = linspace(0,2*np.pi,50) #should be linearArray, or just array
y = sin(x)
clf()
plot(x,y)
xlabel('time') #should be xlabel = 'time'
z = noise + y
plot(x,z,'o')
from scipy.optimize import curve_fit
#curve_fit(f, xdata, ydata, p0=None, sigma=None, **kw)
#Use non-linear least squares to fit a function, f, to data.
#Returns
#-------
#popt : array
#    Optimal values for the parameters so that the sum of the squared error
#    of ``f(xdata, *popt) - ydata`` is minimized
#pcov : 2d array
#    The estimated covariance of popt.  The diagonals provide the variance
#    of the parameter estimate.
sinfunc = lambda x,a,b: a*np.cos(x-b)
fitpars, covmat = curve_fit(sinfunc,x,z)
# The diagonals of the covariance matrix are variances
# variance = standard deviation squared, so we'll take the square roots to get the standard devations!
# You can get the diagonals of a 2D array easily:
variances = covmat.diagonal()
std_devs = np.sqrt(variances)

# Let's plot our best fit, see how well we did
# These two lines are equivalent:
plot(x, sinfunc(x, fitpars[0], fitpars[1]), 'r-')
plot(x, sinfunc(x, *fitpars), 'r-')

#okay... let's try with a different function then...
def func(x, a, b, c):
    return a*np.exp(-b*x) + c

    
popt, pcov = curve_fit(func, x, z)
plot(x, func(x, *popt), 'k-')


```
