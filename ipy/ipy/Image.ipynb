```python
#http://matplotlib.org/users/image_tutorial.html
#import matplotlib.image as mpimg
img = imread('th.png') #img=mpimg.imread('th.png')
lum_img = img[:,:,0]
figsize(12,12)
subplot(1,3,1)
imgplot = plt.imshow(img)
subplot(1,3,2)
imshow(lum_img).set_cmap('gray')
b_n_w = empty(img.shape)
for i in range(len(img)):
    for j in range(len(img[i])):
        avg = sum(img[i][j])/3 
        for k in range(3):
            b_n_w[i][j][k] = avg
        
subplot(1,3,3)

imshow(b_n_w)
simg = b_n_w[:,:,0]
show()
```

```python
tbnw =fft2(simg)
subplot(1,2,1)
imshow(abs(fftshift(tbnw))).set_cmap('Set3')
xlim([100,160])
ylim([70,115])
subplot(1,2,2)
imshow(log(abs(fftshift(tbnw)))).set_cmap('hot')

```

```python
noise = 0.06
sing = simg + noise*randn(simg.shape[0],simg.shape[1])
subplot(1,2,1)
imshow(sing).set_cmap('hot')
subplot(1,2,2)
singt = fft2(sing)
imshow(log(abs(fftshift(singt)))).set_cmap('hot')

```

```python
kx = arange(1,simg.shape[1]+1,1)
ky = arange(1,simg.shape[0]+1,1)
KX,KY = meshgrid(kx,ky)

F = exp(-0.0001*((KX-129)**2+(KY-92)**2))

imshow(F).set_cmap('hot')

from mpl_toolkits.mplot3d import axes3d
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.view_init(azim=50, elev=10)

ax.plot_surface(KX, KY, F)
ax.set_zlim(0, 2)
plt.show()
```

```python
noise = 0.05
sing = simg + noise*randn(simg.shape[0],simg.shape[1])
singt = fft2(sing)
width = 0.000001
F = exp(-width*((KX-simg.shape[0]/2)**2+(KY-simg.shape[1]/1)**2))
singf = F*singt
ss = abs(ifft2(fftshift(singf)))
figsize(15,15)
subplot(131)
imshow(ss).set_cmap('hot')
subplot(132)
imshow(sing).set_cmap('hot')
subplot(133)
imshow(simg).set_cmap('hot')



```
