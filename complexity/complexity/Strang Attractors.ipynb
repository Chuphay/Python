```python
def next_x(old_x, old_y, a):
    return old_y + 1 - a * old_x**2
def next_y(old_x, b):
    return b*old_x
    
```

```python
b = 0.3
a = 1.4
x = [0]
y = [0]
for i in range(1000):
    x.append(next_x(x[i],y[i],a))
    y.append(next_y(x[i],b))
    
```

```python
scatter(x,y)
```
